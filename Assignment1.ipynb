## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import datetime
day1= datetime.date(2023,1,1,)
print('day1:',day1.ctime())
day2= datetime.date(2023,2,10)
print('day2:',day2.ctime())
print('Number of days:',day1-day2)
```
